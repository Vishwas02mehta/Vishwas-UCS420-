```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Q.1 - Part I: Randomized Sales Data Generation (NumPy)
roll_number = 1234
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
columns = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df_sales = pd.DataFrame(sales_data, index=months, columns=columns)

# Q.2 - Part II: Data Manipulation and Analysis (Pandas)
df_sales.head()
df_sales.describe()

total_sales_category = df_sales.sum()
total_sales_month = df_sales.sum(axis=1)

sales_growth = df_sales.pct_change().mean() * 100

df_sales['Total Sales'] = df_sales.sum(axis=1)
df_sales['Growth Rate'] = df_sales['Total Sales'].pct_change() * 100

if roll_number % 2 == 0:
    df_sales['Electronics'] *= 0.9
else:
    df_sales['Clothing'] *= 0.85

df_sales

# Q.3 - Part III: Visualizations (Matplotlib and Seaborn)
plt.figure(figsize=(10, 6))
for column in df_sales.columns[:-2]:
    plt.plot(df_sales.index, df_sales[column], label=column)

plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_sales.iloc[:, :-2])
plt.title('Sales Distribution for Each Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')
plt.grid(True)
plt.show()

# Q.4 - NumPy Array Operations
array = np.array([[1, -2, 3], [-4, 5, -6]])

abs_array = np.abs(array)

percentiles = {
    '25th Percentile (Column)': np.percentile(array, 25, axis=0),
    '50th Percentile (Column)': np.percentile(array, 50, axis=0),
    '75th Percentile (Column)': np.percentile(array, 75, axis=0),
    '25th Percentile (Row)': np.percentile(array, 25, axis=1),
    '50th Percentile (Row)': np.percentile(array, 50, axis=1),
    '75th Percentile (Row)': np.percentile(array, 75, axis=1)
}

mean_values = np.mean(array, axis=0), np.mean(array, axis=1)
median_values = np.median(array, axis=0), np.median(array, axis=1)
std_dev_values = np.std(array, axis=0), np.std(array, axis=1)

abs_array, percentiles, mean_values, median_values, std_dev_values

# Q.5 - Floor, Ceiling, Truncated, and Rounded Values
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])

floor_values = np.floor(a)
ceiling_values = np.ceil(a)
truncated_values = np.trunc(a)
rounded_values = np.round(a)

floor_values, ceiling_values, truncated_values, rounded_values

# Q.6 - Swap Two Elements in a List Using a Temporary Variable
def swap_list_elements(lst, index1, index2):
    temp = lst[index1]
    lst[index1] = lst[index2]
    lst[index2] = temp
    return lst

sample_list = [1, 2, 3, 4, 5]
swap_list_elements(sample_list, 1, 3)

# Q.7 - Swap Two Elements in a Set by Converting it to a List First
def swap_set_elements(s, index1, index2):
    s_list = list(s)
    s_list[index1], s_list[index2] = s_list[index2], s_list[index1]
    return set(s_list)

sample_set = {1, 2, 3, 4, 5}
swap_set_elements(sample_set, 0, 4)

```
